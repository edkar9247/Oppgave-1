```python
#a)oppretter en dictionary med land og deres hovedstad og antall innbyggere i mill. i hovedstaden
data = {"Norge": ["Oslo", 0.634],
        "England": ["London", 8.982],
        "Frankrike": ["Paris", 2.161],
        "Italia": ["Roma", 2.873]}

```

```python
#b)et program som sp√∏r om √• skrive inn et land
land = input("Skriv inn et land: ")
```

```python
#lager et program som skal skrive ut: "London er hovedstaden i England og det er 8.982 mill. innbyggere i London ".
if land in data:
    hovedstad, innbyggere = data[land]
    print(f"{hovedstad} er hovedstaden i {land} og det er {innbyggere} mill. innbyggere i {hovedstad}.")
else:
    print(f"Beklager, jeg har ingen informasjon om {land}.")
```

```python
#c)sjekker land i dictionary
print("N√•v√¶rende land i dictionary:")
for land in data:
    print("-", land)
```

```python
#legge til et nytt land
nytt_land = input("\nSkriv inn et nytt land (som ikke allerede finnes): ")
```

```python
if nytt_land in data:
    print(f"{nytt_land} finnes allerede i dictionary!")
else:
    hovedstad = input(f"Skriv inn hovedstaden i {nytt_land}: ")
    innbyggere = float(input(f"Skriv inn antall innbyggere (i millioner) i {hovedstad}: "))
    
    
    
```

```python
 #legger inn det nye landet i dictionary
data[nytt_land] = [hovedstad, innbyggere]
print("\nDictionaryen er oppdatert!")
```

```python
#
print("\nOppdatert dictionary:")
for land, info in data.items():
    print(f"{land}: Hovedstad = {info[0]}, Innbyggere (mill.) = {info[1]}")
```

```python
#et program med en funksjon som tar a og b som inn-argumenter og som s√• 
#regner ut arealet og ¬´ytre¬ª omkrets til en figur satt sammen av en rettvinklet trekant og en 
#halvsirkel.

import math

def figur(a, b):
    areal_trekant = (a * b) / 2
    areal_halvsirkel = 0.5 * math.pi * (b / 2)**2
    areal = areal_trekant + areal_halvsirkel
    hypotenus = math.sqrt(a**2 + b**2)
    omkrets = a + hypotenus + (math.pi * b) / 2
    return areal, omkrets    
```

```python
# define a and b
a = float(input("Enter the length of side a: "))
b = float(input("Enter the length of side b: "))
```

```python
areal, omkrets = figur(a, b)
```

```python
print(f"\nArealet av figuren er {areal:.3f}")
print(f"Ytre omkrets er {omkrets:.3f}")
```
